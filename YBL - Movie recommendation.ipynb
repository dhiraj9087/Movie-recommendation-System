# Data Preprocessing

```python
df_feature = df[['Movie_Genre','Movie_Keywords','Movie_Tagline','Movie_Cast','Movie_Director']].fillna('')
```

```python
df_feature.shape
```

```python
df_feature
```

```python
X = df_feature['Movie_Genre'] + ' ' + df_feature['Movie_Keywords'] + ' '+ df_feature['Movie_Tagline']+ ' '+df_feature['Movie_Cast']+' '+df_feature['Movie_Director']
```

```python
X
```

```python
X.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
model = TfidfVectorizer()
```

```python
X = model.fit_transform(X)
```

```python
X.shape
```

```python
X
```

```python
print(X)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
Sim_score = cosine_similarity(X)
```

```python
Sim_score
```

```python
Sim_score.shape
```

```python
name=input("enter movie name: ")
```

```python
title_list=df['Movie_Title'].tolist()
```

```python
title_list
```

```python
import difflib
```

```python
movie_recomd = difflib.get_close_matches(name,title_list)
```

```python
print(movie_recomd)
```

```python
close_match = movie_recomd[0]
print(close_match)
```

```python
ind_of_close_match = df[df.Movie_Title==close_match]['Movie_ID'].values[0]
```

```python
ind_of_close_match
```

```python
recomd_score = list(enumerate(Sim_score[ind_of_close_match]))
```

```python
recomd_score
```

```python
len(recomd_score)
```

```python
sorted_sim_movie = sorted(recomd_score, key = lambda x:x[1], reverse=True)
```

```python
sorted_sim_movie
```

```python
print("Top 30 movies for you: \n")
a=0
for movie in sorted_sim_movie:
    index = movie[0]
    title_from_index = df[df.index==index]['Movie_Title'].values[0]
    if a<31:
        print(a,' ',title_from_index)
        a+=1
```

```python
name = input(("Enter your favourite movie name: "))
list_of_all_titles = df['Movie_Title'].tolist()
Find_close_match = difflib.get_close_matches(name,list_of_all_titles)
close_match = Find_close_match[0]
ind_of_movie = df[df.Movie_Title==close_match]['Movie_ID'].values[0]
recomd_score = list(enumerate(Sim_score[ind_of_movie]))
sorted_sim_movie = sorted(recomd_score, key=lambda x:x[1], reverse =True)
print("TOP 10 MOVIES: \n")
a=0
for movie in sorted_sim_movie:
    index = movie[0]
    title_from_index = df[df.index==index]['Movie_Title'].values[0]
    if a<11:
        print(a,'.',title_from_index)
        a+=1
```
